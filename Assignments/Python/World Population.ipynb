# Data Vizualisation

# Pie Chart of World population

```python
y = np.array([0,20,40,60,80,100])
l= ["Density", "Population", "ChangePerc", "NetChange", "Urban", "UrbanPerc"]
c = ['r', 'b', 'g', 'k']

plt.pie(y,labels=l, startangle=90, shadow=True, colors=c, explode=(0,0.5,0,0,0,0))
plt.legend(loc='lower right', title='World population')
plt.show()
```

# Bar graph of Density Vs ChangePerc

```python
plt.figure(figsize=(10,5))
plt.bar(wp["Density"],wp["ChangePerc"])
plt.title("Density Vs ChangePerc",fontsize=2,color="red")
plt.xlabel("Density",fontsize=15)
plt.ylabel("ChangePerc",fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(y='UrbanPerc', data=wp, palette='tab10')
plt.show()
```

# Pie Chart of UrbanPerc

```python
plt.figure(figsize=(10,10))
plt.title('UrbanPerc', fontsize=20)
wp['UrbanPerc'].value_counts().plot.pie(autopct='%1.1f%%, shadow=True')
plt.show()
```

```python
wp.groupby('Density')[['ChangePerc','UrbanPerc']].sum().plot.bar(color=['red','green'], figsize=(8,5))
plt.ylabel('ChangePerc/UrbanPerc')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Density', data=wp, palette='rocket_r',order=wp['Density'].value_counts().index)
plt.xlabel('Density', fontsize=25)
plt.xticks(rotation=70)
plt.show()
```
