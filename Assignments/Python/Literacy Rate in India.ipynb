# Data Visualization

# Pie Chart

```python
y = np.array([0,20,40,60,80,100])
l= ["LR_total-2001", "LR_total_2011", "LR-Rural-2001", "LR_Rural-2011", "LR_Urban_2001", "LR_Urban_2011"]
c = ['r', 'b', 'g', 'k']

plt.pie(y,labels=l, startangle=90, shadow=True, colors=c, explode=(0,0.5,0,0,0,0))
plt.legend(loc='lower right', title='Literacy rate across State')
plt.show()
```

# Literacy rate total in 2001  Vs  Literacy rate total in 2011

```python
plt.figure(figsize=(10,5))
plt.bar(lit["LR_Rural_2011"],lit["LR_total_2001"])
plt.title("LR_Rural_2011 Vs LR_total_2001",fontsize=2,color="red")
plt.xlabel("LR_Rural_2011",fontsize=15)
plt.ylabel("LR_total_2001",fontsize=15)
plt.grid(True)
plt.show()
```
