# Data Visualization

# Pie Chart of Latest Covid -19 in lndia statewise


```python
y = np.array([0,20,40,60,80,100])
l= ["States", "Total Cases", "Active", "Discharged", "Deaths", "Active Ratio"]
c = ['r', 'b', 'g', 'k']

plt.pie(y,labels=l, startangle=90, shadow=True, colors=c, explode=(0,0.5,0,0,0,0))
plt.legend(loc='lower right', title='Covid-19 in statewise')
plt.show()
```
