# datavisualisation

### barplot
in this bar plot we try to visualise the active cases in the states and we can see that the more than 150000 cases are active in the state kerala.

```python
plt.figure(figsize=(10,20))

sns.barplot(x='Active',y='states',data=covid19,palette='tab10')
plt.title('total cases in each states',fontsize=15,color='red') 
plt.xlabel('Active',fontsize=15) 
plt.ylabel('states',fontsize=15) 
plt.show()
```

#### : like the above plot this is also a barplot,in this we compare the number of deaths in each state and we can see that the amount of death is more in the state of maharashtra.

```python
plt.figure(figsize=(10,20))

sns.barplot(x='Deaths',y='states',data=covid19,palette='tab10')
plt.title('deaths according to states',fontsize=15,color='red') 
plt.xlabel('Active',fontsize=15) 
plt.ylabel('states',fontsize=15) 
plt.show()
```

###   stripplot 
in this strip plot we try to visualise the active cases in each state . kerala has the most active cases at this momemnt

```python
sns.set_style('darkgrid')
sns.stripplot(x='Active',y='states',data=covid19,jitter=0.50)
plt.show()
```

###  relational plot
here also we plot the active vs total cases . here we can see that total cases are more in the state maharashtra and more number of active cases is more in kerala.

```python
plt.figure(figsize=(16,18))
sns.relplot(x='Active',y='Total Cases',hue='states',data=covid19)
plt.show()
```

###   lineplot
in this  we try to visualise the most discharged cases in each case and maharashtra have most  number of discharged cases in india


```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Discharged',y='states',data=covid19)
plt.show()
```

### catplot
in this death vs states is to be plotted and in this we can analyse that in the state maharashtra have most number of death cases 

```python
sns.catplot(x='Deaths',y='states',kind='boxen',data=covid19)
plt.show()
```

###  kdeplot
in this kdeplot we try to visualise that the death vs active cases in india

```python
sns.kdeplot(covid19['Deaths'],covid19['Active'],color='hotpink',vertical=True,shade=True)
plt.show()
```

###  barplot
 in this barplot we try to visualise the statewise deaths and active cases. active in green colour and deaths in yellow colour. kerala have the most active cases and maharashtra have most death cases

```python
covid19.groupby('states')[['Active','Deaths']].sum().plot.bar(color=['green','yellow'],figsize=(9,6))
plt.ylabel('Active/Deaths')
plt.show()
```

### pair plot  

```python
sns.pairplot(covid19,diag_kind='kde')
plt.show()
```

###   swarm plot
in this swarm plot we visualise the death ratio of each state . in state punjab have the highest death ratio 

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x='deathratio',y='states',data=covid19,dodge=True,size=10)
plt.show()
```

###  heat map

```python
sns.heatmap(covid19.corr(),cmap='RdBu',annot=True,fmt='.1f',linewidth=1,linecolor='yellow',square=True)
plt.show()
```

###  distplot
in this distplot we try to visualise the deaths comparison with density distribution

```python
sns.distplot(covid19.Deaths,color='hotpink')
plt.show()
```
